```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.array([10,20,30])
y=np.array([40,50,60])
x1=np.array([4,6,8,12])
y1=np.array([22,54,23,2])
plt.plot(x,y)
plt.plot(x1,y1)
plt.show()
```

```python
sizes=np.array([10,50,20,30])
plt.scatter(x,y,color='yellow')
plt.scatter(x1,y1,color='green',s=sizes, alpha=0.8)
plt.show()
```

```python
plt.bar(x1,y1)
plt.show()
```

```python
plt.barh(x1,y1,color='yellow',height=10)
plt.show()
```

```python
plt.bar(x1,y1,color='green',width=2)
plt.show()
```

```python
plt.hist(x1)
plt.show()
```

```python
plt.pie(x1)
plt.show()
```

```python
track_stud=np.array([20,30,40,20,10])
track_name=["Data","Mobile","Web","IoT","CyberSecurity"]
ex=[0.2,0,0,0,0]
col=["Red","Pink",'m','g','b']
plt.pie(track_stud,labels=track_name,startangle=90,explode=ex,shadow=True,colors=col)
plt.legend(title="Tracks")
plt.show()
```
